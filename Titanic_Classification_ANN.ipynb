* NOTE : Exploratory Data Analysis has been already done for the used dataset in another project. Hence it is not shown in this project. This project emphasises only on the implementation of Densely connected Neeural Network to improve classification accuracy.

```python
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.shape , test.shape
```

```python
d = pd.merge(train,test)
```

```python
dataset = train.append(test,ignore_index=True)
```

```python
dataset.shape
```

```python
train_idx = len(train)
test_idx = len(dataset) - len(test)
```

```python
train = dataset[ :train_idx]
test = dataset[test_idx: ]
```

```python
train.shape , test.shape
```

```python
dataset.head()
```

```python
dataset.drop(['PassengerId','Name','Ticket', 'Cabin'],axis = 1, inplace = True)
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
data = dataset
```

```python
def fill_age(col):
    Age= col[0]
    Pclass= col[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return 37
        
        if Pclass==2:
            return 29
        
        if Pclass == 3:
            return 24
        
    else:
        return Age
```

```python
data['Age'] = data[['Age','Pclass']].apply(fill_age,axis=1)
```

```python
data['Fare']=data['Fare'].fillna(data['Fare'].mean())

data['Embarked'] = data['Embarked'].fillna('S')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['FamilySize'] = data.SibSp + data.Parch + 1
```

```python
data.head()
```

```python
data.Sex = data.Sex.map({'male':0, 'female':1})
```

```python
data.head()
```

```python
pclass_dummies = pd.get_dummies(data['Pclass'],drop_first=True)
```

```python
embarked_dummies = pd.get_dummies(data['Embarked'],drop_first=True)
```

```python
data_dummies = pd.concat([data,pclass_dummies,embarked_dummies],axis = 1)
```

```python
data_dummies.head()
```

```python
data_dummies.head()
```

```python
train_data = data_dummies[ :train_idx]
test_data = data_dummies[test_idx: ]
```

```python
train_data.head()
```

```python
train_data.shape
```

```python
test_data.shape
```

```python
test_data.head()
```

```python
X_train = train_data.drop('Survived',axis=1)
y_train = train_data.Survived
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
X_test = test_data.drop('Survived',axis=1)
```

```python
X_test.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
y_train = y_train.values
```

```python
y_train
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
import keras 
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(units=6, kernel_initializer='uniform', activation='relu', input_dim=10))
```

```python
classifier.add(Dense(units=6, kernel_initializer='uniform', activation='relu'))
```

```python
classifier.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))
```

```python
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
classifier.fit(X_train, y_train, batch_size=10, epochs = 150)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred = (y_pred>0.5)*1
```

```python
y_pred
```

```python
test.head()
```

```python
submission_ann = pd.DataFrame({
        "PassengerId": test["PassengerId"],
        "Survived": y_pred.flatten()
    })
submission_ann.to_csv("submission_ann.csv", index = False)
```
